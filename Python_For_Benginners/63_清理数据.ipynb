```python
import pandas as pd
```

```python
name = pd.Series(["小陈", "小李", "小王", "小张", "小赵", "小周"],
                 index=["1", "002", "003", "004", "005", "006"])
gender = pd.Series(["女", "女", "男", "男", "女", "男"],
                index=["006", "005", "004", "003", "002", "1"])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7],
                   index=["1", "002", "003", "004", "005"])
```

```python
student = pd.DataFrame({'姓名':name,'性别':gender,'身高':height})
student
```

### rename方法改索引index和列名columns

### rename里面传入的是一个字典前面的键是原来的索引或列名后面的值是修改后的索引或列名

```python
student.rename(index={'1':'001'})
```

### rename方法不会改变原始df

```python
student
```

### 重新赋值or传入参数inplace

```python
student.rename(index={'1':'001'},inplace=True)
student
```

```python
student=student.rename(index={'1':'001'})
student
```

#### rename还可以传入函数或方法作为参数

### 把某列设为索引

```python
student.set_index('姓名',inplace=True)
```

### 重设索引

```python
student.reset_index()
```
