```python
import pandas as pd
```

```python
name = pd.Series(["小陈", "小李", "小王", "小张", "小赵", "小周"],
                 index=["001", "002", "003", "004", "005", "006"])
gender = pd.Series(["女", "女", "男", "男", "女", "男"],
                index=["006", "005", "004", "003", "002", "001"])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7],
                   index=["001", "002", "003", "004", "005"])
```

```python
student = pd.DataFrame({'姓名':name,'性别':gender,'身高':height})
student
```

## 评估丢失数据用info方法（注意不要忘记括号）

```python
student.info()
```

### 还可以用isnull方法

```python
student.isnull()
```

### 再用sum方法求和

```python
student.isnull().sum()
```

#### True和False用于数学计算是分别为1，0

#### 可以计算单独一列也可以直接计算每列

```python
student['身高'].isnull().sum()
```

### 提取变量缺失的行1.没有sum2.是单独一列

```python
student[student['身高'].isnull()]
```

## 评估重复数据

### 用duplicated在Series中重复数据为True在DateFrame中只有两行完全相同才为True

```python
student.duplicated()
```

### 若要同时查看几个变量是否重复可以给duplicated传入可选参数subset赋值给一个列表

```python
student.duplicated(subset=['姓名','身高'])
```

### 一样的方法提取行

```python
student[student.duplicated(subset=['姓名','身高'])]
```
