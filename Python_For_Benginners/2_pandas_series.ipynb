```python
import pandas as pd
```

```python
s1 = pd.Series([1,4,2,3,5],
               index=[1,3,5,7,9])
```

```python
s2 = pd.Series([8,1,7,3,9],
              index=[1,2,3,5,10])
```

```python
s1 + s2
```

```python
s1.add(s2,fill_value=0)
```

```python
s1.sub(s2,fill_value=0)
```

```python
s1.div(s2,fill_value=0)
```

```python
s1.mul(s2,fill_value=0)
```

```python
print(s1.max())
print(s1.min())
print(s1.sum())
print(s1.mean())
```
