```python
import pandas as pd
```

```python
name = pd.Series(["小/陈", "小/李", "小/王", "小/张", "小/赵", "小/周"],
                 index=["1", "002", "003", "004", "005", "006"])
gender = pd.Series(["女", "女", "男", "男", "女", "男"],
                index=["006", "005", "004", "003", "002", "1"])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7],
                   index=["1", "002", "003", "004", "005"])
```

```python
student = pd.DataFrame({'姓名':name,'性别':gender,'身高':height})
student
```

#### 对数据进行转置

```python
student.T
```

#### 对列进行拆分用split方法

```python
student['姓名']
```

#### 传出一个列表

```python
student['姓名'].str.split('/')
```

#### 加上expand参数

```python
student['姓名'].str.split('/',expand=True)
```

#### 添加新的列 若添加的列名已经存在则更新，若不存在则添加新列

```python
student[['新1','新2']]=student['姓名'].str.split('/',expand=True)
student
```

#### 删除原来的列

```python
student=student.drop('姓名',axis=1)
student
```

#### 合并列str.cat

```python
student['新1'].str.cat(student['新2'])
```

#### 用sep来指定拼接的分隔符

```python
student['新1'].str.cat(student['新2'],sep='/')
```

#### 把宽数据转化为长数据（把一部分列名的值转化为变量的值）

#### 可以用melt方法可以传入参数                                                  1.要处理的df                                                                     2.di_vars赋值我不为改变的列                                                     3.var_name赋值为包含原本列名值的新列名                                         4.value_name赋值为转换后df中包含原本变量值的新列列名
